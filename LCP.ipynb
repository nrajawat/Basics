```python
#climbing stairs

class Solution:
    def climbStairs(self, n: int) -> int:
        def fib(n):
            if n <= 0:
                return 0
            if n == 1:
                return 1

            elif n == 2:
                return 2
        
            else:
                return fib(n-1) + fib(n - 2)

        return fib(n)
                

# Create an instance of the Solution class
solution = Solution()

# Example usage:
n = 5
result = solution.climbStairs(n)
print(f"The number of distinct ways to climb {n} steps is: {result}")

```

```python
# Add two binary numbers

class Solution:
    def addBinary(self, a: str, b: str) -> str:
        c = 0
        res = [ ] 

        max_len = max(len(a), len(b))
        a = a.zfill(max_len)
        b = b.zfill(max_len)

        for i in range(max_len -1, -1, -1):
            bit_sum = int(a[i]) + int(b[i]) + c
            res.append(str(bit_sum % 2))
            c = bit_sum // 2

        if c:
            res.append(str(c))
        return ' '.join(res[::-1])
                                      

```

```python
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        carry = 0
        result = []

        # Ensure both binary strings have the same length by padding with zeros
        max_len = max(len(a), len(b))
        a = a.zfill(max_len)
        b = b.zfill(max_len)

        # Perform binary addition from right to left
        for i in range(max_len - 1, -1, -1):
            bit_sum = int(a[i]) + int(b[i]) + carry
            result.append(str(bit_sum % 2))
            carry = bit_sum // 2

        # If there's a carry after the loop, add it to the result
        if carry:
            result.append(str(carry))

        # Reverse the result list and convert it to a binary string
        return ''.join(result[::-1])

# Example usage:
sol = Solution()
binary1 = "1010"
binary2 = "1101"
sum_binary = sol.addBinary(binary1, binary2)
print(sum_binary)  # Output: "11011"
```

```python
#oneplus

from typing import List

class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        # Convert the list of digits into a single number
        num = 0
        for digit in digits:
            num = num * 10 + digit

        # Add 1 to the number
        num += 1

        # Convert the number back to a list of digits
        result = []
        while num > 0:
            result.append(num % 10)
            num //= 10

        # Reverse the list to get the correct order of digits
        result.reverse()

        return result

# Example usage:
input_digits = [1, 2, 3]
sol = Solution()
result = sol.plusOne(input_digits)
print(result)  # Output: [1, 2, 4] (123 + 1 = 124)
```

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if not needle:
            return 0

        n = len(haystack)
        m = len(needle)

        for i in range(n - m + 1):
            if haystack[i:i + m] == needle:
                return i

        return -1

# Example usage:
haystack_str = "hello, world"
needle_str = "world"
sol = Solution()
result = sol.strStr(haystack_str, needle_str)
print(result)  # Output: 7 (needle "world" is found at index 7 in the haystack string)
```

```python
from typing import List

class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        write_index = 0

        # Iterate through the array.
        for num in nums:
            if num != val:
                nums[write_index] = num
                write_index += 1

        # Resize the array to remove the occurrences of 'val'.
        del nums[write_index:]

        return write_index

# Example usage:
input_array = [3, 2, 2, 3, 1, 4, 2, 5]
val_to_remove = 2
sol = Solution()
result = sol.removeElement(input_array, val_to_remove)
print(result)  # Output: 5 (modified array will be [3, 3, 1, 4, 5])
```

```python
from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if not nums:
            return 0

        # Initialize two pointers, one for iterating through the array and one for placing unique elements.
        write_index = 1

        # Iterate through the array starting from the second element.
        for i in range(1, len(nums)):
            if nums[i] != nums[i - 1]:  # Check if the current element is different from the previous one.
                nums[write_index] = nums[i]
                write_index += 1

        return write_index

# Example usage:
input_array = [1, 1, 2, 2, 3, 3, 4, 4, 5]
sol = Solution()
result = sol.removeDuplicates(input_array)
print(result)  # Output: 5 (modified array will be [1, 2, 3, 4, 5, 3, 4, 4, 5])
```

```python
from typing import List

class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        if not strs:
            return ""

        # Find the minimum length among the input strings
        min_len = min(len(s) for s in strs)

        # Iterate over the characters at each position of the first string
        prefix = ""
        for i in range(min_len):
            char = strs[0][i]
            for s in strs:
                if s[i] != char:
                    return prefix
            prefix += char

        return prefix

# Example usage:
input_strings = ["flower", "flow", "flight"]
sol = Solution()
print(sol.longestCommonPrefix(input_strings))  # Output: "fl"
```

```python
def longestCommonPrefix(strs):
    if not strs:
        return ""

    # Sort the list of strings to make sure the shortest and longest strings are at the ends.
    strs.sort()

    # Find the minimum length of the first and last strings (the shortest and longest strings).
    min_len = min(len(strs[0]), len(strs[-1]))

    # Compare characters of the shortest and longest strings.
    prefix = ""
    for i in range(min_len):
        if strs[0][i] == strs[-1][i]:
            prefix += strs[0][i]
        else:
            break

    return prefix

# Example usage:
input_strings = ["flower", "flow", "flight"]
print(longestCommonPrefix(input_strings))  # Output: "fl"
```
