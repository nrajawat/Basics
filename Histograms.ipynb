```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame with the fish data
data = pd.DataFrame({
    'Width': [21, 20, 20, 19, 17, 16, 18, 18, 15, 17],
    'Lightness': [3, 4, 8, 5, 9, 5, 6, 7, 7, 3],
    'Label': ['Salmon', 'Salmon', 'Sea bass', 'Salmon', 'Sea bass', 'Sea bass', 'Salmon', 'Salmon', 'Sea bass', 'Sea bass']
})

# Define the bins for width and lightness
width_bins = [14, 16, 18, 20, 22]
lightness_bins = [2, 4, 6, 8, 10]

# Separate the data based on the fish labels
salmon_data = data[data['Label'] == 'Salmon']
sea_bass_data = data[data['Label'] == 'Sea bass']

# Create the width histogram
plt.figure(figsize=(8, 6))
plt.hist([salmon_data['Width'], sea_bass_data['Width']], bins=width_bins, color=['blue', 'green'], label=['Salmon', 'Sea bass'])
plt.xlabel('Width')
plt.ylabel('Count')
plt.title('Width Histogram')
plt.legend()
plt.show()

# Create the lightness histogram
plt.figure(figsize=(8, 6))
plt.hist([salmon_data['Lightness'], sea_bass_data['Lightness']], bins=lightness_bins, color=['blue', 'green'], label=['Salmon', 'Sea bass'])
plt.xlabel('Lightness')
plt.ylabel('Count')
plt.title('Lightness Histogram')
plt.legend()
plt.show()
```
